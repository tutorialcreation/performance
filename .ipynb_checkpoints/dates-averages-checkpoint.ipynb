```python
# first function for grouping deadlines
def deadlines_group(request):
    member_subtasks=SubTask.objects.filter(member_assigned=request.user)
    
    grouped_deadlines=[]
    for subtask in member_subtasks:
        grouped_deadlines.append(subtask.task_due_date.date())
    set(grouped_deadlines)
    
    # for finding averages
    averages={}
    for subtask in martin_subtasks:
        due_date=subtask.task_due_date.date()
        if averages.get(due_date,''):
            averages[due_date].append(subtask.subtask_deadline_score)
        else:
            averages[due_date]=[subtask.subtask_deadline_score]
            
#     import statistics
    result={}
    for average in averages:
        avg=[item for item in averages[average] if item]
        result.update({average:statistics.mean(avg)}) if avg else result.update({average:avg})
    #     statistics.mean([5,6,7])
    return JsonResponse(data={
        'labels':[date_.strftime("%d %b") for date_ in list(result.keys())],
        'data':list(result.values())
    })
    
```
